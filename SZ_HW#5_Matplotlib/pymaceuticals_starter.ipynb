```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_df = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_df = pd.read_csv(clinical_trial_data_to_load)

# mouse_drug_df.head()
# clinical_trial_df.head()

# Combine the data into a single dataset
combined_df = pd.merge(clinical_trial_df, mouse_drug_df, on="Mouse ID", how='left')

# Display the data table for preview
combined_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumor_volume_mean = combined_df.groupby(["Drug", "Timepoint"]).mean()
# Convert to DataFrame
tumor_volume_mean_df = tumor_volume_mean.drop(["Metastatic Sites"], axis=1)
tumor_volume_mean_df = tumor_volume_mean_df.reset_index()
# Preview DataFrame
tumor_volume_mean_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_volume_SE = combined_df.groupby(["Drug", "Timepoint"]).sem()
# Convert to DataFrame
tumor_volume_SE_df = tumor_volume_SE.drop(["Metastatic Sites","Mouse ID"], axis=1)
tumor_volume_SE_df = tumor_volume_SE_df.reset_index()
# Preview DataFrame
tumor_volume_SE_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
Mungled_Data_tumor = tumor_volume_mean_df.pivot(index = "Timepoint", columns = "Drug", values = "Tumor Volume (mm3)") 
# Preview that Reformatting worked
Mungled_Data_tumor.head()
```

```python
# Generate the Plot (with Error Bars)
Capomulin_SE = tumor_volume_SE_df.loc[tumor_volume_SE_df["Drug"] == "Capomulin", "Tumor Volume (mm3)"]
Infubinol_SE = tumor_volume_SE_df.loc[tumor_volume_SE_df["Drug"] == "Infubinol", "Tumor Volume (mm3)"]
Ketapril_SE  = tumor_volume_SE_df.loc[tumor_volume_SE_df["Drug"] == "Ketapril",  "Tumor Volume (mm3)"]
Placebo_SE   = tumor_volume_SE_df.loc[tumor_volume_SE_df["Drug"] == "Placebo",   "Tumor Volume (mm3)"]

Time = np.arange(0, 50, 5)

fig, ax = plt.subplots()
plt.errorbar(Time, Mungled_Data_tumor["Capomulin"], yerr=Capomulin_SE, label="Capomulin", marker="o", c="r", ls='--')
plt.errorbar(Time, Mungled_Data_tumor["Infubinol"], yerr=Infubinol_SE, label="Infubinol", marker="^", c="b", ls='--')
plt.errorbar(Time, Mungled_Data_tumor["Ketapril"],  yerr=Ketapril_SE,  label="Ketapril",  marker="s", c="g", ls='--')
plt.errorbar(Time, Mungled_Data_tumor["Placebo"],   yerr=Placebo_SE,   label="Placebo",   marker="D", c="k", ls='--')

plt.title("Tumor Response to Treatment")
plt.legend()
plt.grid()

ax.set_xlim(Time[0]-3, Time[-1]+3)
ax.set_xlabel("Time (Days)")
ax.set_ylabel("Tumor Volume (mm$^3$)")

# Save the Figure
plt.savefig("Images/Tumor_Response_to_Treatment.png")
# Show the Figure
plt.show()
```

```python
# Generate the Plot (with Error Bars)
Capomulin_SE = Metsite_SE_df.loc[Metsite_SE_df["Drug"] == "Capomulin", "Metastatic Sites"]
Infubinol_SE = Metsite_SE_df.loc[Metsite_SE_df["Drug"] == "Infubinol", "Metastatic Sites"]
Ketapril_SE  = Metsite_SE_df.loc[Metsite_SE_df["Drug"] == "Ketapril",  "Metastatic Sites"]
Placebo_SE   = Metsite_SE_df.loc[Metsite_SE_df["Drug"] == "Placebo",   "Metastatic Sites"]

Time = np.arange(0, 50, 5)

fig, ax = plt.subplots()
plt.errorbar(Time, Mungled_Data_Meta["Capomulin"], yerr=Capomulin_SE, label="Capomulin", marker="o", c="r", ls='--')
plt.errorbar(Time, Mungled_Data_Meta["Infubinol"], yerr=Infubinol_SE, label="Infubinol", marker="^", c="b", ls='--')
plt.errorbar(Time, Mungled_Data_Meta["Ketapril"],  yerr=Ketapril_SE,  label="Ketapril",  marker="s", c="g", ls='--')
plt.errorbar(Time, Mungled_Data_Meta["Placebo"],   yerr=Placebo_SE,   label="Placebo",   marker="D", c="k", ls='--')

plt.title("Metastatic Spread During Treatment")
plt.legend()
plt.grid()

ax.set_xlim(Time[0]-3, Time[-1]+3)
ax.set_xlabel("Treatment Duration (Days)")
ax.set_ylabel("Met Sites")

# Save the Figure
plt.savefig("Images/Metastatic_Response_to_Treatment.png")
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mice_count = combined_df.groupby(["Drug", "Timepoint"])["Mouse ID"].count()
# Convert to DataFrame
mice_count_df = pd.DataFrame(mice_count)
mice_count_df = mice_count_df.reset_index()
# Preview DataFrame
mice_count_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
Mice_Data = mice_count_df.pivot(index = "Timepoint", columns = "Drug", values = "Mouse ID") 
# Preview the Data Frame
Mice_Data.head()
```

```python
# Generate the Plot (Accounting for percentages)
Total_Mice = 25
Mice_perc1 = (Mice_Data["Capomulin"] / Total_Mice)*100
Mice_perc2 = (Mice_Data["Infubinol"] / Total_Mice)*100
Mice_perc3 = (Mice_Data["Ketapril"]  / Total_Mice)*100
Mice_perc4 = (Mice_Data["Placebo"]   / Total_Mice)*100

Time = np.arange(0, 50, 5)

fig, ax = plt.subplots()
plt.plot(Time, Mice_perc1, label="Capomulin", marker="o", c="r", ls='--')
plt.plot(Time, Mice_perc2, label="Infubinol", marker="^", c="b", ls='--')
plt.plot(Time, Mice_perc3, label="Ketapril",  marker="s", c="g", ls='--')
plt.plot(Time, Mice_perc4, label="Placebo",   marker="D", c="k", ls='--')

plt.title("Survival During Treatment")
plt.legend()
plt.grid()

ax.set_xlim(Time[0]-3, Time[-1]+3)
ax.set_xlabel("Time (Days)")
ax.set_ylabel("Survival Rate (%)")

# Save the Figure
plt.savefig("Images/Survival_Rates.png")
# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
tumor_volume_initial = Mungled_Data_tumor.iloc[0,:]
tumor_volume_final   = Mungled_Data_tumor.iloc[-1,:]
tumor_volume_change_perc = (tumor_volume_final -tumor_volume_initial) /tumor_volume_initial *100
# Display the data to confirm
tumor_volume_change_perc
```

```python
# Store all Relevant Percent Changes into a Tuple
change_perc_list_of_tuple = list(zip(tumor_volume_change_perc.index, tumor_volume_change_perc))
change_perc_list_of_tuple
```

```python
# Splice the data between passing and failing drugs
drug_pass = []
drug_fail = []
drug_pass_index = []
drug_fail_index = []

for ind in range(len(change_perc_list_of_tuple)):
    change = change_perc_list_of_tuple[ind][1]
    change_ind = change_perc_list_of_tuple[ind][0]
    if change < 0 :
        drug_pass.append(change)
        drug_pass_index.append(change_ind)
    else :
        drug_fail.append(change)
        drug_fail_index.append(change_ind)
        
passing_drugs = list(zip(drug_pass_index, drug_pass))
failing_drugs = list(zip(drug_fail_index, drug_fail))
```

```python
plot_var = np.round([tumor_volume_change_perc["Capomulin"], tumor_volume_change_perc["Infubinol"],
                     tumor_volume_change_perc["Ketapril"], tumor_volume_change_perc["Placebo"]])
x_values = range(len(plot_var))

fig, ax = plt.subplots()
bar = ax.bar(x_values, plot_var, color='r', alpha=0.8, align="edge", width = -1)
bar[0].set_color('g')

plt.title("Tumor Change Over 45 Day Treatment")

# Orient widths. Add labels, tick marks, etc. 
Xlabels = ["Capomulin", "Infubinol", "Ketapril", "Placebo"]
ax.set_xticks(x_values)
ax.set_xticklabels(Xlabels)

Ylabels = np.arange(-20, 80, 20)
ax.set_yticks(Ylabels )
ax.set_yticklabels(Ylabels )
ax.set_ylabel("% Tumor Volume Change")
ax.set_ylim(-30, 70)

plt.grid()

# Use functions to label the percentages of changes
def label(bars):
    for i in bars:
        height = i.get_height()
        ax.text(i.get_x() + i.get_width()/2, .2*height, "%d" %int(height)+ "%",
                ha='center', va='top', color="white")

# Call functions to implement the function calls
label(bar)

# Save the Figure
plt.savefig("Images/Percentage_Tumor_Volume_Change.png")

# Show the Figure
plt.show()
```
